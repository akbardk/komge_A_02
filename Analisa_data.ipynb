```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8))
plt.plot(data['rata_rata'], label='Rata-Rata')
plt.plot(data['terendah'], label='Terendah')
plt.plot(data['tertinggi'], label='Tertinggi')

plt.xlabel('Jumlah Data')
plt.ylabel('Nilai UN')
plt.title('Grafik Hasil UN SMP,SMA, dan SMK Tahun 2012-2012 di DKI Jakarta')
plt.legend()
plt.show()
```
